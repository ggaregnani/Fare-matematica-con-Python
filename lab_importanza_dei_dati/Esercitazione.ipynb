# Matrice di confusione

Importiamo le librerie

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Inserire qui i risultati del test

```python
v1 = 6  # Il modello ha detto 1 ed era 1
f1 = 9 # Il modello ha detto 1 ma era 2
v2 = 3  # Il modello ha detto 2 ed era 2
f2 = 2 # Il modello ha detto 1 ma era 1
```

Creiamo la tabella

```python
cm = np.array([[v1, f1],
               [f1, v2]])
```

Visualizziamo la matrice di confusione

```python
labels = ['1', '2']
fig, ax = plt.subplots(figsize=(6, 5))

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax,
            xticklabels=labels, yticklabels=labels,
            annot_kws={"size": 16}, cbar=False)

ax.set_title("Matrice di Confusione", fontsize=14, pad=20)
ax.set_xlabel("Predizione dell'IA", fontsize=12)
ax.set_ylabel("Dato Reale (Target)", fontsize=12)

plt.show()
```

Calcoliamo l'accuratezza

```python
print("accuracy:", (v1+v2)/(v2+v1+f1+f2))
```
