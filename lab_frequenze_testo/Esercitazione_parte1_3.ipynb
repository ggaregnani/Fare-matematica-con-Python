# Parte 1

Importa le librerie necessarie.

```python
import re
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Lettura dei dati

Importa i dati da un archivio
dei contenuti di Wikipedia e messo a disposizione sulla piattaforma
[Hugging Face](https://huggingface.co/datasets/Salesforce/wikitext) (Merity et al., 2016).

```python
from datasets import load_dataset

dataset = load_dataset("wikitext", "wikitext-103-raw-v1")
```

Utilizza prima solo i dati del dataset *test*. In un secondo momento puoi ripetere l'esercitazione con il dataset *train*, cambiando il nome tra parentesi quadre.

```python
data = dataset["test"]
testo = " ".join(data["text"])
print("Lunghezza testo:", len(testo))
```

## Tokenizzazione

Crea una lista delle parole presenti.

```python
tokens = re.findall(r"\b\w+\b", testo.lower())
tokens[:10]
```

Quante parole totali sono presenti?

```python
print(len(tokens))
```

## Analisi delle frequenze

Per ogni *parola type* conta le *parole token* e riporta la frequenza assoluta in una tabella ordinata dalla parola più frequente alla meno frequente. Aggiungi una colonna con la frequenza relativa e una con la posizione della *parola type* nella tabella ordinata, cioè il *rango*.

```python

frequenze = Counter(tokens)

df = (
    pd.DataFrame(frequenze.items(), columns=["type", "frequenza_assoluta"])
    .sort_values(by="frequenza_assoluta", ascending=False)
    .reset_index(drop=True)
)

df["frequenza_relativa"] = df.frequenza_assoluta/len(tokens)
df["rango"] = df.index + 1

df.head()
```

## La legge nascosta

Per trovare la legge nascosta calcola il logaritmo della frequenza assoluta e del rango.

```python
log_rango = np.log(df['rango'])
log_freq = np.log(df["frequenza_relativa"])
```

Riporta nel piano cartesiano i logaritmi delle frequenze relative di ogni parola type in funzione dei logaritmi del rango.

```python
plt.figure()
plt.scatter(log_rango, log_freq)
plt.xlabel("log(Rango)")
plt.ylabel("log(Frequenza)")
plt.ylim(top=0)
plt.title("Legge di Zipf")
plt.grid()
plt.show()
```
