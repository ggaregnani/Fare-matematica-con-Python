# Analisi statistica di un testo



Importa le librerie necessarie

```python
import re
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importazione dati

Importa i dati da un archivio
dei contenuti di Wikipedia e messo a disposizione sulla piattaforma
[Hugging Face](https://huggingface.co/datasets/Salesforce/wikitext) (Merity et al., 2016).

```python
from datasets import load_dataset

dataset = load_dataset("wikitext", "wikitext-103-raw-v1")
```

Leggi prima solo i dati del dataset test. In un secondo momento puoi riprovare a fare l'esercitazione con il dataset train.

```python
data = dataset["test"]
testo = " ".join(data["text"])
print("Lunghezza testo:", len(testo))
```

## Tokenizzazione



```python
tokens = re.findall(r"\b\w+\b", testo.lower())
tokens[:10]
```

## Analisi delle frequenze


```python
# Calcolo la lunghezza del corpus
l_corpus= len(tokens)
print(l_corpus)
```

```python
# Per ogni parola type conto le parole token e riporto la frequenza assoluta in una tabella
frequenze = Counter(tokens)

df = (
    pd.DataFrame(frequenze.items(), columns=["type", "frequenza"])
    .sort_values(by="frequenza", ascending=False)
    .reset_index(drop=True)
)

df.head()
```

```python
# Aggiungo il rango

df["rango"] = df.index + 1
```

```python
# Calcolo la frequenza relativa

df["frequenza_relativa"] = df.frequenza/l_corpus
df.head()
```

## La legge nascosta

Per trovare la legge nascosta calcola i logaritmi e ricaviamo dal grafico l'equazione della retta.


```python
log_rango = np.log(df['rango'])
log_freq = np.log(df["frequenza_relativa"])
```

Riporta nel piano cartesiano le frequenze relative di ogni parola type in funzione del suo rango in un piano log-log.

```python
plt.figure()
plt.scatter(log_rango, log_freq)
plt.xlabel("log(Rango)")
plt.ylabel("log(Frequenza)")
plt.title("Legge di Zipf")
plt.grid()
plt.show()
```

Determina m e q della retta rappresentata

```python
m=-1.17
q=-0.90
```

```python
log_y=m*log_rango+q
```

```python
plt.figure()
plt.scatter(log_rango, log_freq)
plt.plot(log_rango, log_y, linestyle="--", label="Interpolazione")
plt.xlabel("log(Rango)")
plt.ylabel("log(Frequenza)")
plt.title("Legge di Zipf")
plt.grid()
plt.show()
```

```python
m=(-1+5)/(0-4)
print(m)
```

```python
log_ranks = np.log(df['rango'])
log_freq = np.log(df["frequenza_relativa"])

# fit lineare
slope, intercept = np.polyfit(log_ranks, log_freq, 1)
print(slope, intercept)
```

### Grafico loglog

```python
plt.figure()
plt.loglog(df["rango"], df["frequenza_relativa"])
plt.xlabel("Rango")
plt.ylabel("Frequenza")
plt.title("Legge di Zipf")
plt.grid()
plt.show()
```
